Data analysis of 1000 IMDB movies (2006-2016) and finding the trends of what people like and genre they prefer to watch

Dataset  contains all the data of IMDB from the year 2006 to 2016. The era where some of the greatest movies were made

```python
#Downloading all the required documentation.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

# Data cleaing and processing

```python
df = pd.read_csv("/content/IMDB-Movie-Data.csv")
df
```

```python
df.info()
```

We can say that data set contains about 12 columns and we can see that **there are some null values** as well in Revenue and Metascore

```python
df.rename(columns={'Runtime (Minutes)':'Runtime' ,
                   'Revenue (Millions)': 'Revenue'},inplace=True, errors='raise')
```

```python
df.isnull().sum()
```

Here we cans see that revenue has aboout 128 null and Metascore has about 64 null.

```python
df.Revenue.fillna(value = 0,inplace = True )
df.Metascore.fillna(value = 0,inplace = True )
```

```python
df.isna().sum()
```

Here we replace all the NAN values with 0 since  revenue and Metascore is thye row specific and each  movie  has its own specific value for those and these values have no relation to the nearby values

```python
df['Genre'] = df.Genre.str.split(',')
df['Actors'] = df.Actors.str.split(',')
```

Converting the string into array so that thry can be easily accessible for using array funcion for splitting them for visualisation

Deleting the non required column so that we can work on it more easily

# Data analysis

```python
fig = px.box(df, x="Year", y="Revenue",title='Revenue of top movies by year')
fig.show()
```

From above graph we can see average income per year and we can draw the following interferance in following points-
NOTE- since most of the values where NAN converted to 0 there might be some changes.
1. Maximum income was done in year 2015 of 936.6 million followed by movie in 2009 with revenue of 760 million
2. most of income values fall between about 80 -100
3. most movies in 2016 has made 3.1 million


```python
dfg = df.explode("Genre")
```

```python
fig = px.histogram(dfg, x="Genre",color = "Year",title='Movies in each  genre')
fig.show()
```

From above we can say that drama is the most liked genre followed by action  and then comedy. Drama has 500 plus,Action has 300 and comedy about 270.
Musical is the least liked with about only 5 likes


```python
dfg = df.explode("Genre")
fig = px.bar(dfg, x="Genre",y = "Revenue",color = "Year",title='Total revenue earned by top 1000 movies by each genre')
fig.show()
```

Highest Revenue is earned by adventure movies followed by action and then drama.Adventure has earned  about 42k with action  about 37k.
Musical was produced least so has earned the least

```python
fig = px.histogram(dfg, x="Director",color = "Year",title='Directors with thier movie releases')
fig.show()
```

We can say that most top  movies are released by Ridney Scott.

```python
pop = dfg[dfg.Director == "Ridley Scott"].Genre.value_counts()
cop = ["Drama","Adventure","Action","Sci-Fi","Crime","Romance","Mystery","Biography","Thriller","Comedy"]
fig = px.pie(df, values=pop, names=cop, title='Ridley Scott most produced genre')
fig.show()
```

Ridley Scott mostly produces  Dramas followed by adventure.

```python
dfa =df.explode('Actors')
```

Now we are done with genre let let us move to  actors analysis

```python
fig = px.histogram(dfa, x="Actors",color = "Year")
fig.show()
```

We can see that Mark Wahlberg & Christian Bale  has made the most top movies so we can assume that he  is a favourate actor

```python

a = dfg.groupby(['Director']).agg({'Revenue':'mean','Metascore':'mean'})
x = a.sort_values(by=['Revenue','Metascore'], ascending = False).head(10)
x.reset_index(level=0, inplace=True)

fig = go.Figure()
fig.add_trace(go.Bar(
    x=x.Director,
    y=x.Revenue,
    name='Revenue',
    marker_color='blue'
))
fig.add_trace(go.Bar(
    x=x.Director,
    y=x.Metascore,
    name='Metascore',
    marker_color='green'
))
fig.update_layout(barmode='group', xaxis_tickangle=-45,title='Revenue and metascore of the Directors')
fig.show()
```

We can say that James Cameron has the higest revenue and has made the most liked movies,followed by Colin Trevorrow
